# Downloading

## Getting the lists

### Characters

```python
baseurl = "https://rickandmorty.fandom.com/api.php?"
action = "action=query"
theList = "list=categorymembers"
cmtitle = "cmtitle=Category:Characters"
cmlimit = "cmlimit=700"
content = "prop=revisions&rvprop=content"
dataformat = "format=json"

query = "%s%s&%s&%s&%s&%s&%s" % (baseurl, action, theList, cmtitle, cmlimit, content, dataformat)
print(query)
```

```python
response = urllib.request.urlopen(query)
source = response.read()
```

```python
wikiPageName = []

jsonObj = json.loads(source)
for i in jsonObj['query']['categorymembers']:
    wikiPageName.append(i['title'])
```

```python
action = "action=query"
revision = "prop=revisions"
content = "rvprop=content"
dataformat = "format=json"

for name in wikiPageName:
    print(name)
    
    # Parsing name for url
    urlName = urllib.parse.quote(name, safe='')
        
    # Building the query
    title = "titles=" + urlName
    query = "%s%s&%s&%s&%s&%s" % (baseurl, action, revision, title, content, dataformat)
    
    # Calling API
    response = urllib.request.urlopen(query)
    source = response.read()

    # Loading Json
    jsonObj = json.loads(source)
    
    # Getting page ID
    pageid = list(jsonObj['query']['pages'].keys())[0]
    
    # Getting content
    fileContent = jsonObj['query']['pages'][pageid]['revisions']
    
    # Converting to a utf-8 string
    utfContent = str(str(fileContent).encode("utf-8"))
    
    # Writing in files
    file = open(name + ".json", "w")    
    file.write(utfContent) 
    file.close()
```

### Locations

```python
cmtitle = "cmtitle=Category:Locations"

query = "%s%s&%s&%s&%s&%s&%s" % (baseurl, action, theList, cmtitle, cmlimit, content, dataformat)

response = urllib.request.urlopen(query)
source = response.read()

locations = []

jsonObj = json.loads(source)
for i in jsonObj['query']['categorymembers']:
    locations.append(i['title'])

locations = [x for x in locations if "Category" not in x]
locations
```

```python
for name in locations:
    print(name)
    
    # Parsing name for url
    urlName = urllib.parse.quote(name, safe='')
        
    # Building the query
    title = "titles=" + urlName
    query = "%s%s&%s&%s&%s&%s" % (baseurl, action, revision, title, content, dataformat)
    
    # Calling API
    response = urllib.request.urlopen(query)
    source = response.read()

    # Loading Json
    jsonObj = json.loads(source)
    
    # Getting page ID
    pageid = list(jsonObj['query']['pages'].keys())[0]
    
    # Getting content
    fileContent = jsonObj['query']['pages'][pageid]['revisions']
    
    # Converting to a utf-8 string
    utfContent = str(str(fileContent).encode("utf-8"))
    
    # Writing in files
    
    # : and / are not allowed in filenames so had to be replaced by - and %
    fileName = name.replace(":","-")
    fileName = fileName.replace("\\","%")
    file = open(fileName + ".json", "w")    
    file.write(utfContent)
    file.close()
```

```python
path = os.getcwd() + "/Characters/"

characters = []
G = nx.DiGraph()

# Collecting names
for file in os.listdir(path):
    name = file.split(".json")[0]
    name = name.replace("%","\\")
    name = name.replace(":","-")
    characters.append(name)
    
# Creating DataFrame
# df = pd.DataFrame(characters, columns = ['Name','Specie','Status'])
df = pd.DataFrame(characters, columns = ['Name'])

for name in characters:
    G.add_node(name)

edges = []

for file in os.listdir(path):
    valid_links = []

    # Reading file
    f = io.open(path + file,'r', encoding = 'utf-8').read()
    
    # Getting links
    links = re.findall("\[\[(.*?)\]\]",f)
        
    # Sorting links
    for link in links:
        for c_name in characters:
            if (c_name in link) and ('Category:' not in link) and ('Image:' not in link):
                valid_links.append(c_name)
    
    edges.append(valid_links)
```

```python
print(nx.info(G))
```

```python
nodes = list(G.nodes)

for j in range(0,len(nodes)):
    for k in edges[j]:
        G.add_edge(nodes[j],k)
```
